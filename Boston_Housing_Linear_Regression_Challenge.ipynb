```
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_csv('/content/gdrive/My Drive/bostonhousing.csv')
df
```

```
import seaborn as sn
sn.heatmap(df.corr(), annot=True)
```

```
df.plot(kind='scatter',y='medv',x='rm')
```

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
 
scaled = scaler.fit_transform(df[['rm', 'lstat', 'ptratio']])
 
 
df_scaled = pd.DataFrame(data=scaled, columns=['rm_scaled', 'lstat_scaled', 'ptratio_scaled'])
 
```

```
df = df.join(df_scaled)

from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error

LR=LinearRegression()
 
X = df[['rm', 'lstat', 'ptratio']]
y = df['medv']
 
LR.fit(X, y)
 
df['prediction'] = LR.predict(X)
np.sqrt(mean_squared_error(df['medv'],df['prediction']))
```
